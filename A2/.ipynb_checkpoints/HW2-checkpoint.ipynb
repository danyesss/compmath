````python
def GetPlot(mask, i, j):
    global axs;
    
    x = data[mask]["x"]
    y = data[mask]["y"]
    
    axs[i][j].set_title("Dx={:.0f}, Dy={:.2f}, r={:.2f}".format(np.var(x), np.var(y), np.corrcoef(x, y)[0,1]))
    axs[i][j].scatter(x, y, c="r", s=10)
    axs[i][j].set_xlim(3, 20)
    axs[i][j].set_ylim(2, 14)
    
    axs[i][j].minorticks_on()
    axs[i][j].grid(b=True, which='major', color='k', linestyle='-', alpha=0.2)
    axs[i][j].grid(b=True, which='minor', color='k', linestyle='--', alpha=0.1)
````

````python
fig, axs = plt.subplots(2, 2, figsize=(12, 6), dpi=100)
fig.tight_layout(pad=2)

GetPlot(m1,0,0)
GetPlot(m2,0,1)
GetPlot(m3,1,0)
GetPlot(m4,1,1)
````

**Task 3**

На лекции обсуждалось, что метод наименьших квадратов — это способ поставить задачу о решении *переопределенной*
системы $Xw = y$, которая имеет явный ответ, выражающийся через левую псевдообратную матрицу для $X$. Для
*недоопределенной* системы $Xw = y$ (имеющей бесконечно много решений) можно поставить задачу о поиске решения с минимальной $l_2$ -нормой весов $\lVert w \rVert = w^T w$. Решите такую задачу и покажите, что ответ выражается через правую псевдообратную матрицу для $X$. Считайте, что прямоугольная матрица $X$ имеет полный ранг (максимально
возможный).

**Решение**

$$Xw = y$$
В данном случае *loss function* из полноты ранга обнуляется на некотором множестве весов $\{w\}$

Задачу можно решить с помощью метода неопр. коэффициентом Лагранжа:

$$\mathcal{L} = w^T w + \lambda^T (Xw - y)$$

$$\partial_w\mathcal{L} = 2w^T + \lambda^T X = 0$$

$$\partial_\lambda\mathcal{L} = Xw - y = 0$$

Откуда получаем, что

$$\lambda = -2(X X^T)^{-1} y$$

$$w = \underbrace{X^T(X X^T)^{-1}}_{\text{right pseudoinverse}} y$$

````python
df0arr = np.asarray(df0)
# df0arr = np.transpose(df0)
df0means = df0arr.mean(axis=0);
print(' Mean Brightness Temperature: ', round(df0means[0]+273.15,1), '±', round(df0means[1],1), '\n',
      'Mean Thermodynamical Temperature: ', round(df0means[9],1), '±', round(df0means[10],1))
print(' Eye\'s error : ±', round(df0means[9]-df0means[0]-273.15))
````

## Lamp Temperature graduation

````python
x = [i for i in df1arr[0] if np.isnan(i) == False]
y = [i for i in df1arr[1] if np.isnan(i) == False]

[P,V] = np.polyfit(x, y, 1)

x_fit = x;
x_fit = np.append(x_fit,800.0)
x_fit = np.append(x_fit,2100.0)
````

````python
x_fit = np.sort(x_fit)
y_fit = [element * P + V for element in x_fit]
````

**Fig 1 : Graduation curve for Lamp temperature**

````python
plt.plot(x,y,'o', color = 'black', label = "grad scatter")
plt.plot(x_fit, y_fit, label = "grad scatter fit" )

plt.grid(which='major')

plt.xlabel(r'$T_{bright}$')
plt.ylabel(r'$T_{real}$')
plt.show()
````
