```python
def gauss_map(x, a, b, A, B):
    return (x * (A - B) + a * B - A * b )/(a - b)
```

```python
a, b = -1, 1
A, B = 0, np.pi/4
x1, x2 = -1/np.sqrt(3), 1/np.sqrt(3)

X1 = gauss_map(x1, a, b, A, B)
X2 = gauss_map(x2, a, b, A, B)

C1 = np.sin(2*X1)/(np.pi/8) * (B-A)/2
C2 = np.sin(2*X2)/(np.pi/8) * (B-A)/2

print(X1, X2)
print(C1, C2)
```

```python
class LagrangeInterpolator:
    def __init__(self, xk, yk):
        self.xk = np.asarray(xk, dtype=float)
        self.yk = np.asarray(yk, dtype=float)
        
    def __call__(self, x):
        val = 0
        N = (self.xk).shape[0]
        for k in range(N):
            xs_k = np.hstack([self.xk[:k],self.xk[k+1:]]) # все x_i без x_k
            Xs = np.array([x for i in range(N-1)]) # охапка из x
            ck = np.prod((Xs - xs_k) / (self.xk[k] - xs_k))
            val += ck * self.yk[k]
        return val
        raise NotImplementedError()
```

```python
xs = np.array([4, 8, 9])
ys = np.sqrt(xs**3)
ys
```

```python
xs = np.array([4, 8, 9])
ys = np.sqrt(xs**3)


# написанный варинт
LI = LagrangeInterpolator(xs, ys)

# библиотечный вариант
LL = sp.interpolate.lagrange(xs, ys);

x0 = np.linspace(0, 16, 100)
y0 = np.sqrt(x0**3)

plt.scatter(xs, ys, s=50, color="k")
plt.plot(x0, y0, linewidth=10, alpha=0.25, label="$\sqrt{x^3}$")
plt.plot(x0, [LI(x) for x in x0], "--", color="k")
plt.plot()
plt.xlim(4, 16)
plt.ylim(0, 70)
plt.legend()
plt.show()
```

```python
from scipy.sparse import linalg
```

```python
def generate_matrix_A(N, eps):
    h = 1/N
    
    d1 = eps/(h**2) * np.ones(N-1)
    d0 = (7 -  (4 * eps)/(h**2))*np.ones(N)
          # diagonals = 
    return sp.sparse.diags([d1, d0, d1], [-1, 0, 1]).toarray()
```

```python
A = generate_matrix_A(10, 1)
eigs = sp.sparse.linalg.eigs(A, return_eigenvectors=False)
```
